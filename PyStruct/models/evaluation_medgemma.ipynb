Runs the current MedGemma baseline on `data/labels.csv`, then saves predictions, metrics, classification report, and the confusion matrix.

```python
from pathlib import Path
import sys

project_root = Path.cwd()
if project_root.name == "models":
    project_root = project_root.parent

if str(project_root) not in sys.path:
    sys.path.insert(0, str(project_root))

project_root
```

```python
from collections import Counter
from models.test import read_images

dataset = list(read_images())
label_counts = Counter(label for _, label in dataset)

len(dataset), dict(label_counts)
```

```python
from models.test import run_predictions

results = run_predictions(batch_size=5)
results[:2]
```

```python
from models.test import (
    build_metrics,
    save_confusion_matrix,
    save_metrics,
    save_predictions,
    CONFUSION_MATRIX_PATH,
    METRICS_PATH,
    PREDICTIONS_PATH,
)

save_predictions(results)
metrics = build_metrics(results)
save_metrics(metrics)
save_confusion_matrix(metrics)

print(f"Accuracy: {metrics['accuracy']:.2f}")
print(f"Balanced accuracy: {metrics['balanced_accuracy']:.2f}")
print(metrics["classification_report_text"])
print(f"Predictions: {PREDICTIONS_PATH}")
print(f"Metrics: {METRICS_PATH}")
print(f"Confusion matrix: {CONFUSION_MATRIX_PATH}")
```
